```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data=sns.load_dataset("tips")
print(data)
sns.boxplot(data=data,x="tip")#iss mai hm dekhte hai ki esa konsa data hai jo mean value se extreme bahar hai
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data=sns.load_dataset("tips")
print(data)
sns.boxplot(data=data,x="sex",y="tip",orient="vertical")#iss mai hm dekhte hai ki esa konsa data hai jo mean value se extreme bahar hai
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
data=sns.load_dataset("tips")
print(data)
sns.boxplot(data=data,x="day",y="tip",orient="vertical",fliersize=5)#iss mai hm dekhte hai ki esa konsa data hai jo mean value se extreme bahar hai
plt.show()
```
